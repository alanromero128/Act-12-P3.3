## Las variables
1. X --> cantidad de lapices
2. Pag --> pago

## Las formulas
si x>=1000
$ PAG = X*0.85$
si no
$ PAG = X*0.90$

## Diagrama de flujo LucidChart

<img src="https://raw.githubusercontent.com/alanromero128/Act-12-P3.3/refs/heads/main/DF%20if%20else%20pago%20lapices.png" width="300" height="500" />



```python
# para cuando es verdadero (si)
# Calculando pago de lapices  usando if else
# Datos de entrada X
# mostrar PAG
X = int(input("Ingrese la cantidad de lapices: "))
if  X>=1000:
    PAG = X*0.85
else:
    PAG = X*0.9


print("el pago es :", PAG)
print("Fin del programa")
print("Programa realizado por Alan Romero NL = 47")
```

```python
# para cuando es falso (no)
# Calculando pago de lapices  usando if else
# Datos de entrada X
# mostrar PAG
X = int(input("Ingrese la cantidad de lapices: "))
if  X>=1000:
    PAG = X*0.85
else:
    PAG = X*0.9


print("el pago es :", PAG)
print("Fin del programa")
print("Programa realizado por Alan Romero NL = 47")
```
